### Step 9. Visualize the Crime Rate

```python
dataset['Violent_Crime'] = dataset['Murder']+dataset['Forcible_Rape']+dataset['Robbery']+dataset['Aggravated_assault']
dataset['Property_Crime']= dataset['Burglary']+dataset['Larceny_Theft']+dataset['Vehicle_Theft']

print(dataset)
```

```python
sns.scatterplot(x = dataset.Year, y = dataset.Violent_Crime, color = 'green')
sns.scatterplot(x = dataset.Year, y = dataset.Property_Crime, color = 'red')
plt.title("Crime Rate in Each Year", size = 15)
plt.xlabel('Year', size = 15)
plt.ylabel('Number of crimes', size=15)
```

**The Red plot points in the Visualization depicts that property crime is the most frequent one.
